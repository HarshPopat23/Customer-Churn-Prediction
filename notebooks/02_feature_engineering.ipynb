```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler

```

```python
df = pd.read_csv('../data/telco_churn.csv')
df_cleaned = df.copy()

df_cleaned.head()
```

```python
df_cleaned['is_male'] = df_cleaned['gender'].map({'Male': 1, 'Female': 0})
df_cleaned['is_Partner'] = df_cleaned['Partner'].map({'Yes': 1, 'No': 0})
df_cleaned['is_Dependents'] = df_cleaned['Dependents'].map({'Yes': 1, 'No': 0})
df_cleaned['is_PhoneService'] = df_cleaned['PhoneService'].map({'Yes': 1, 'No': 0})
df_cleaned['is_PaperlessBilling'] = df_cleaned['PaperlessBilling'].map({'Yes': 1, 'No': 0})
df_cleaned['is_Churn'] = df_cleaned['Churn'].map({'Yes': 1, 'No': 0})
```

```python
df_cleaned['TotalCharges'] = pd.to_numeric(
    df_cleaned['TotalCharges'], errors='coerce'
)

df_cleaned['TotalCharges'].fillna(
    df_cleaned['TotalCharges'].median(), inplace=True
)
```

```python
scaler = StandardScaler()

num_cols = ['tenure', 'MonthlyCharges', 'TotalCharges']
df_cleaned[num_cols] = scaler.fit_transform(df_cleaned[num_cols])
```

```python
df_cleaned = pd.get_dummies(
    df_cleaned,
    columns=[
        'PaymentMethod',
        'Contract',
        'MultipleLines',
        'InternetService',
        'OnlineSecurity',
        'OnlineBackup',
        'DeviceProtection',
        'TechSupport',
        'StreamingTV',
        'StreamingMovies'
    ],
    drop_first=True
)
```

```python
df_cleaned = df_cleaned.drop(columns=[
    'customerID',
    'gender',
    'Partner',
    'Dependents',
    'PhoneService',
    'PaperlessBilling',
    'Churn'
])
```

```python
df_cleaned.select_dtypes(include='object').columns
```

```python
X = df_cleaned.drop(columns=['is_Churn'])
y = df_cleaned['is_Churn']
```

```python
X.to_csv('../data/X_processed.csv', index=False)
y.to_csv('../data/y_processed.csv', index=False)

print("✅ Processed data saved successfully.")
```
